```python
import pandas as pd
import matplotlib.pyplot as plt
import regex as re
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
TN_county_pop = pd.read_csv('../dq3-earthquakes-down-to-earth/county_pop.csv')
```

```python
TN_county_pop.drop_duplicates(subset='County', inplace=True)
TN_county_pop.sort_values(by='2019POP',ascending=False,inplace=True)
TN_county_pop['County'] = TN_county_pop['County'].str.replace('County','')
TN_county_pop
```

```python
sns.set_palette('PuRd')
plt.figure(figsize=(20,7))
ax = sns.barplot(data=TN_county_pop,x='County',y='2019POP')
ax.set(ylabel='2019 Population')
plt.xticks(rotation=45,ha='center')
plt.show()
#i want to edit the counties so they are only on the easter and western side as well as fix the wonky centering of the xtick labels
```
